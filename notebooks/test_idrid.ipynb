# Test IDRiD Dataloader

This notebook verifies the dataset loader for the
**IDRiD** dataset.

```python
import os
from config.constants import IDRID_PATH

from dataloader.idrid import IDRiDModule

from utils.transforms import eval_transform
print("Python path:", os.getcwd())
root = os.environ.get("IDRID_PATH", IDRID_PATH)
print("Using root:", root)
# assert os.path.exists(root), f"Dataset root not found: {root}"
transform = eval_transform(224)
dm = IDRiDModule(root=root, transform=transform, batch_size=8)
dm.setup(stage="fit")
print("Dataset ready")
try:
    loader = dm.val_dataloader()
except Exception as exc:
    raise
print("len(val_ds)=", len(dm.val_ds))
batch = next(iter(loader))
imgs, labels, paths = batch
print("batch images shape:", imgs.shape)
print("batch labels shape:", labels.shape)
print("sample paths:", paths[:3])
```
